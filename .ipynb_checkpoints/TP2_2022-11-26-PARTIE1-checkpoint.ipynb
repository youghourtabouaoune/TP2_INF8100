```python
from bs4 import BeautifulSoup
import requests
import json
import re
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from IPython.display import HTML
import hvplot.pandas
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score


# Convertir la taille du terrain en nombre entier pied carré. par exemple 450.52x220.05 sera 450.52 X 220.05 = 99136.926
# Appel: dfData["Taille terrain"]=dfData["Taille terrain"].apply(lambda x: convertTailleTerrainToNumber(x))
def convertTailleTerrainToPIcarre(n):
    try:
        float(n)
        return n
    except ValueError:
        try:
            splitValues = n.upper().split('X')
            newValue=pd.to_numeric(splitValues[0])  * pd.to_numeric(splitValues[1])
            return newValue   
        except ValueError:
            # Ce n'est pas le bon format, on remplace la valeur par nulle
            return np.nan 
    
#Obtenir une liste des liens par numéro de la page.
def getListLinksPageNo(pageNumber):
    url=("https://duproprio.com/fr/rechercher/liste?search=true&regions%5B0%5D=1&regions%5B1%5D=10&regions%5B2%5D=11&regions%5B3%5D=114&regions%5B4%5D=115&regions%5B5%5D=12&regions%5B6%5D=13&regions%5B7%5D=14&regions%5B8%5D=15&regions%5B9%5D=16&regions%5B10%5D=17&regions%5B11%5D=2&regions%5B12%5D=257&regions%5B13%5D=3&regions%5B14%5D=31&regions%5B15%5D=4&regions%5B16%5D=5&regions%5B17%5D=6&regions%5B18%5D=7&regions%5B19%5D=8&regions%5B20%5D=9&with_builders=1&parent=1&pageNumber="+str(pageNumber)+"&sort=price")
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html5lib')
    
    #Liste de tous les liens via : <div class="search-results-listings-list__container">
    allLinks=soup.find_all("div", class_="search-results-listings-list__container")

    linksList = []
    for i in range(len(allLinks)):
        linksList.append(allLinks[i].a['href'])

    return linksList

# Obtenir le nombre maximale disponible par site
def getMaxPageNoAvailable():
    noPage = 1
    pagePrecedent=getListLinksPageNo(noPage)
    pageSuivante=""

    while pagePrecedent != pageSuivante: 
        noPage += 1
        pageSuivante = getListLinksPageNo(noPage)

    return noPage        

def getAdresse(soup):
   
    strNomRue=""
    try:
        strNomRueTag=soup.find("div", class_="listing-location__address") 
        strNomRue=strNomRueTag.h1.text
    except:
        strNomRue=""
    
    return strNomRue


def getPrix(soup):
   
    try:
        prixdemande=-1
        prixTag=soup.find("div", class_="listing-price__amount")  
        prixdemande=prixTag.text.strip().replace(chr(160), "")

    except:
        prixdemande=-1
    
    return prixdemande

def getAssurance(soup):
   
    try:
        montant=-1
        itemsNamesTag=soup.find_all("div", class_="mortgage-data__table__row__item--name") 
        itemsValuesTag=soup.find_all("div", class_="mortgage-data__table__row__item--yearly-costs") 

        for i in range(len(itemsNamesTag)):
            if (itemsNamesTag[i].text.strip() == "Assurances"):
                montant=itemsValuesTag[i].text.strip().replace(chr(160), "")
                break
    except:
        montant=-1
    
    return montant


def getElectricite(soup):
   
    try:
        montant=-1
        itemsNamesTag=soup.find_all("div", class_="mortgage-data__table__row__item--name") 
        itemsValuesTag=soup.find_all("div", class_="mortgage-data__table__row__item--yearly-costs") 

        for i in range(len(itemsNamesTag)):
            if (itemsNamesTag[i].text.strip() == "Électricité"):
                montant=itemsValuesTag[i].text.strip().replace(chr(160), "")
                break
    except:
        montant=-1
    
    return montant

def getTaxeMunicipale(soup):
   
    try:
        montant=-1
        itemsNamesTag=soup.find_all("div", class_="mortgage-data__table__row__item--name") 
        itemsValuesTag=soup.find_all("div", class_="mortgage-data__table__row__item--yearly-costs") 

        for i in range(len(itemsNamesTag)):
            if (itemsNamesTag[i].text.strip() == "Taxes municipales"):
                montant=itemsValuesTag[i].text.strip().replace(chr(160), "")
                break
    except:
        montant=-1
    
    return montant


def getTaxeScolaire(soup):
   
    try:
        montant=-1
        itemsNamesTag=soup.find_all("div", class_="mortgage-data__table__row__item--name") 
        itemsValuesTag=soup.find_all("div", class_="mortgage-data__table__row__item--yearly-costs") 

        montant=-1
        for i in range(len(itemsNamesTag)):
            if (itemsNamesTag[i].text.strip() == "Taxes scolaires"):
                montant=itemsValuesTag[i].text.strip().replace(chr(160), "")
                break
    except:
        montant=-1
    
    return montant


def getVille(soup):
    try:
        strTemp=""
        data = soup.find('script', text=re.compile("SingleFamilyResidence"))
        json_object = json.loads(data.contents[0])
        strTemp=json_object['address']["addressLocality"]
        
    except:
        strTemp=""
    
    return strTemp.strip()

def getRegion(soup):
    try:
        strTemp=soup.find('div', class_="listing-location__address").find_all("span")
        strTemp=strTemp[1].text
        
    except:
        strTemp=""
    
    return strTemp.strip()


def getNombreChambre(soup):
   
    try:
        itemsNamesTag=soup.find_all("span", class_="listing-main-characteristics__title") 
        itemsValuesTag=soup.find_all("span", class_="listing-main-characteristics__number") 

        nombre=-1
        for i in range(len(itemsNamesTag)):
              if (itemsNamesTag[i].text.strip().__contains__("chambres")):
                nombre=itemsValuesTag[i].text.strip()
                break
    except:
        nombre=-1
    
    return nombre



def getNombreSalleBain(soup):
   
    try:
        itemsNamesTag=soup.find_all("span", class_="listing-main-characteristics__title") 
        itemsValuesTag=soup.find_all("span", class_="listing-main-characteristics__number") 

        nombre=-1
        for i in range(len(itemsNamesTag)):
              if (itemsNamesTag[i].text.strip().__contains__("de bain")):
                nombre=itemsValuesTag[i].text.strip()
                break
    except:
        nombre=-1
    
    return nombre

def getNombreSalleEau(soup):
   
    try:
        itemsNamesTag=soup.find_all("span", class_="listing-main-characteristics__title") 
        itemsValuesTag=soup.find_all("span", class_="listing-main-characteristics__number") 

        nombre=-1
        for i in range(len(itemsNamesTag)):
              if (itemsNamesTag[i].text.strip().__contains__("salle d’eau")):
                nombre=itemsValuesTag[i].text.strip()
                break
    except:
        nombre=-1
    
    return nombre


def getNombreEtage(soup):
   
    try:
        itemsNamesTag=soup.find_all("span", class_="listing-main-characteristics__title") 
        itemsValuesTag=soup.find_all("span", class_="listing-main-characteristics__number") 

        nombre=-1
        for i in range(len(itemsNamesTag)):
            if (itemsNamesTag[i].text.strip().__contains__("étage")):
                nombre=itemsValuesTag[i].text.strip()
                break
    except:
        nombre=-1
    
    return nombre



def getAireHabitable(soup):
   
    try:
        itemsNamesTag=soup.find_all("div", class_="listing-main-characteristics__item-dimensions") 
        itemValuesTag=soup.find_all("span", class_="listing-main-characteristics__number listing-main-characteristics__number--dimensions") 

        nombre=-1
        for i in range(len(itemsNamesTag)):
              if (itemsNamesTag[i].span.text.strip().__contains__("habitable")):
                taille=itemValuesTag[i].text.strip()
                nombre=taille[:taille.index('pi')].replace(" ","")
                break
    except:
        nombre=-1
    
    return nombre



def getTailleTerrain(soup):
   
    try:
        itemsNamesTag=soup.find_all("div", class_="listing-main-characteristics__item-dimensions") 
        itemValuesTag=soup.find_all("span", class_="listing-main-characteristics__number listing-main-characteristics__number--dimensions") 

        nombre=-1
        for i in range(len(itemsNamesTag)):
              if (itemsNamesTag[i].span.text.strip().__contains__("terrain")):
                taille=itemValuesTag[i].text.strip()
                nombre=taille[:taille.index('pi')].replace(" ","")
                break
    except:
        nombre=-1
    
    return nombre


def moindre_carres(x, y):
    beta_1 = np.corrcoef(x, y)[0,1] * np.std(y) / np.std(x)
    beta_0 = np.mean(y) - beta_1 * np.mean(x)
    return beta_0, beta_1

def moindre_carresMultiple(X, y):
    X_T = np.transpose(X) 
    X_bar = X.mean(axis=0)
    y_bar = y.mean()
    Beta = np.linalg.inv(X_T @ (X - X_bar)) @ (X_T @ (y - y_bar)) 
    Beta_0 =  y_bar - X_bar @ Beta 
    return Beta_0, Beta
```

```python
###############################################################
#1. PARTIE 1 :  4.1 COLLECTE DE DONNÉES 
###############################################################
# Extraire tout l’ensemble des annonces dans un fichier .csv à remettre. 
#Les colonnes de fichier csv sont

noPage = 1
pageSuivante=getListLinksPageNo(1)
pagePrecedente=""
firstTime=True
columns = ["Adresse", "Prix","Ville","Région","Chambres","Salles de bain", "Salles d'eau", "Étages", "Aire habitable", "Taille terrain","Taxes municipales","Taxes scolaires","Électricité","Assurances"]

while pagePrecedente != pageSuivante: 
        donnees=[]
        pagePrecedente = pageSuivante
        for link in pagePrecedente:
            url=( link + "#description")
            html = requests.get(url).text
            soup = BeautifulSoup(html, 'html5lib')
            nombreChambre=getNombreChambre(soup)
            nombreSalleBain=getNombreSalleBain(soup)
            nombreSalleEau=getNombreSalleEau(soup)
            nombreEtage=getNombreEtage(soup)
            TailleTerrain=getTailleTerrain(soup)
            AireHabitable=getAireHabitable(soup)

            url=( link + "#couts")
            html = requests.get(url).text
            soup = BeautifulSoup(html, 'html5lib')
            adresse=getAdresse(soup)
            prix=getPrix(soup) 
            ville=getVille(soup)
            region=getRegion(soup)
            electricite=getElectricite(soup)
            taxeMun=getTaxeMunicipale(soup)
            taxescolaire=getTaxeScolaire(soup)
            assurance=getAssurance(soup)

            donnees.append([adresse, prix, ville, region, nombreChambre,nombreSalleBain,nombreSalleEau,nombreEtage,AireHabitable, TailleTerrain, taxeMun, taxescolaire,electricite,assurance  ])

        if (firstTime):
            pd.DataFrame(donnees, columns=columns).to_csv("duproprio.csv", mode='a', encoding='utf_8', index=False) 
            firstTime = False
        else:
            pd.DataFrame(donnees, columns=columns).to_csv("duproprio.csv", mode='a', encoding='utf_8', index=False, header=False)    

        print("Numéro de la pages traité est: ",noPage)
        noPage += 1
        pageSuivante = getListLinksPageNo(noPage)


print("Nombre de total de pages traité est: ",noPage)
```

```python

####################################################################################
##### 2. PARTIE 1 :  4.2 NETTOYAGE ET EXPLORATION DES DONNÉES (4PTS)  ############
####################################################################################
missing_values = ["n/a", "na", "--","NaN", -1]
dfData = pd.read_csv("duproprio.csv", na_values=missing_values)

print ("\n2.1 Combien ya-t-il de valeurs manquantes dans chaque colonne de votre jeude données?")
nombreAvantNettoyage=dfData.shape[0]

print("Les valeurs manquantes dans chaque colonne sont: \n", dfData.isnull().sum())
print("\nNombre total de valeurs manquantes: ", dfData.isnull().sum().sum())

print("\n2.3 Quel est le type (inféré par pandas) de données de chaque colonne ?\n")
print(dfData.dtypes)

###################################################### 2.4 ##################
print("\n2.4 Nettoyer vos données : correction d’erreurs, traitement de valeurs manquantes s’il y a lieu, correction du type des données.")
# Supprime les lignes où il n'y a pas de valeur pour la colonne "Aire habitable".
dfData.dropna(axis=0, subset=["Aire habitable" ], inplace=True)
dfData.dropna(axis=0, subset=["Prix"], inplace=True)

print("\nRemplacer les valeurs manquantes par une valeur fixe zero et autres valeurs nécessaires.")
dfData["Chambres"].fillna(0, inplace=True)
dfData["Salles de bain"].fillna(0, inplace=True)
dfData["Salles d'eau"].fillna(0, inplace=True)
dfData["Étages"].fillna(0, inplace=True)

print ("\nRemplace le caracère '&#039;' par un apostrophe.")
dfData["Ville"]=dfData["Ville"].str.replace("&#039;", "'")

# Enlever les virgules entre les chiffres et
# Convertir la taille du terrain en nombre pi carré
dfData["Taille terrain"]=dfData["Taille terrain"].str.replace(",", "")           
dfData["Taille terrain"]=dfData["Taille terrain"].apply(lambda x: convertTailleTerrainToPIcarre(x))

# Conserve seulement les chiffres entre 0 et 9 
dfData["Aire habitable"] = pd.to_numeric(dfData["Aire habitable"].str.replace("[^0-9]", "", regex=True)).fillna(0)
dfData["Taille terrain"] = pd.to_numeric(dfData["Taille terrain"].str.replace("[^0-9]", "", regex=True)).fillna(0)

# Conserve le signe négatif (-), le point (.) et les chiffres entre 0 et 9
dfData['Prix'] = pd.to_numeric(dfData['Prix'].str.replace("[^-.0-9]", "", regex=True)).fillna(0)
dfData['Taxes municipales'] = pd.to_numeric(dfData['Taxes municipales'].str.replace("[^-.0-9]", "", regex=True)).fillna(0)
dfData['Taxes scolaires'] = pd.to_numeric(dfData['Taxes scolaires'].str.replace("[^-.0-9]", "", regex=True)).fillna(0)
dfData['Électricité'] = pd.to_numeric(dfData['Électricité'].str.replace("[^-.0-9]", "", regex=True)).fillna(0)
dfData['Assurances'] = pd.to_numeric(dfData['Assurances'].str.replace("[^-.0-9]", "", regex=True)).fillna(0)


print("\nSupprimer les maisons coütées 1$");
dfData.drop(dfData[dfData['Prix'] <= 1.0].index, inplace = True)

print("\nSupprimer les maisons n'ont pas de taxes municipale et taxes scolaires");
dfData.drop(dfData[(dfData['Taxes municipales'] == 0) & (dfData['Taxes scolaires'] == 0)].index, inplace = True)

print("\nSupprimer les maisons n'ont pas de la taille du terrain");
dfData.drop(dfData[(dfData['Taille terrain'] == 0)].index, inplace = True)

print("\nSupprimer les maisons n'ont pas de la chambre et les salles d'eau.");
dfData.drop(dfData[(dfData["Chambres"] == 0) & (dfData["Salles d'eau"] == 0)].index, inplace = True)
print("\n");
print(dfData.describe)

print("\nCorriger les types de données")
typesDonnees = {"Aire habitable":"int64","Taille terrain":"int64","Chambres":"int64","Salles de bain":"int64","Salles d'eau":"int64","Étages":"int64","Prix":"float64","Électricité":"float64","Taxes municipales":"float","Taxes scolaires":"float","Assurances":"float64"}        
dfData = dfData.astype(typesDonnees)
print(dfData.dtypes)

print("\nNombre d'enregistrement avant de nettoyage est: ",nombreAvantNettoyage)
print("\nNombre d'enregistrement après de nettoyage est: ",dfData.shape[0])
print("\nEnregistrer les données propres dans un fichier CSV 'duproprioFinale.csv' pour faire la validation.")
dfData.to_csv("duproprioFinale.csv",index=False) 

###################################################### 2.5 ##################
print("\n2.5 Quel est le prix moyen des maisons (au moins 1 chambre et 1 salle de bain) sur l’île de Montréal ? À Laval ? Dans les laurentides ?")
dfPrixMontreal = dfData[ dfData["Ville"].str.upper().str.contains('MONTRÉAL') == True ]
dfPrixLaval = dfData[ dfData["Ville"].str.upper().str.contains('LAVAL') == True ]
dfPrixLaurentides = dfData[ dfData["Ville"].str.upper().str.contains('LAURENTIDES') == True ]

dfPrixMoyenMontreal=dfPrixMontreal['Prix'].mean()
dfPrixMoyenLaval=dfPrixLaval['Prix'].mean()
dfPrixMoyenLaurentides=dfPrixLaurentides['Prix'].mean()

print("Le prix moyen des maisons sur l’île de Montréal est ", dfPrixMoyenMontreal)
print("Le prix moyen des maisons à Laval est ", dfPrixMoyenLaval)
print("Le prix moyen des maisons dans les Laurentides est ", dfPrixMoyenLaurentides)
print()
print()

###################################################### 2.6 ##################
print("\n2.6 Dans quelle ville de Montréal/l’Ile les maisons (au moins 1 chambre et 1 salle de bain) coûtent le moins chers ?")
dfPrixMontreal=dfPrixMontreal[['Ville', 'Prix', 'Chambres']]
dfPrixMontreal=dfPrixMontreal.sort_values(by='Prix')
print(dfPrixMontreal.head(1))

###################################################### 2.7 ##################
print("\n2.7 Pour chaque région, afficher le prix de l’item (annonce) le plus élevé et la ville où l’item se situe. Ici on ne fait pas de différence si c’est un condo/appartement, maison, terrain vide, etc. A quel région/ville revient la palme d’or de l’item le plus cher ? Donner toutes les caractéristiques (valeurs de toutes les colonnes) de cet item.")
dfPlusCher=dfData.sort_values(by='Prix', ascending=False).head(1)
print("\tRégion: ",dfPlusCher["Région"].to_string(index=False))
print("\tAdresse: ",dfPlusCher["Adresse"].to_string(index=False))
print("\tPrix: ",dfPlusCher["Prix"].to_string(index=False))
print("\tVille: ",dfPlusCher["Ville"].to_string(index=False))
print("\tChambres: ",dfPlusCher["Chambres"].to_string(index=False))
print("\tSalles de bain: ",dfPlusCher["Salles de bain"].to_string(index=False))
print("\tSalles d'eau: ",dfPlusCher["Salles d'eau"].to_string(index=False))
print("\tÉtages: ",dfPlusCher["Étages"].to_string(index=False))
print("\tAire habitable: ",dfPlusCher["Aire habitable"].to_string(index=False))
print("\tTaille terrain: ",dfPlusCher["Taille terrain"].to_string(index=False))
print("\tTaxes municipales: ",dfPlusCher["Taxes municipales"].to_string(index=False))
print("\tTaxes scolaires: ",dfPlusCher["Taxes scolaires"].to_string(index=False))
print("\tÉlectricité: ",dfPlusCher["Électricité"].to_string(index=False))
print("\tAssurances: ",dfPlusCher["Assurances"].to_string(index=False))
```
